KMeans and KNN are distinct in their purposes and learning types. KMeans is an unsupervised learning algorithm used for clustering. It organizes data points into a set number of clusters based on their features, without requiring labeled data. It identifies patterns and structures in the data itself. On the other hand, KNN (K Nearest Neighbors) is a supervised learning algorithm used mainly for classification, and occasionally for regression. It predicts the label of a new data point by considering the labels of the 'K' nearest data points in the training set, thus requiring labeled data to understand the relationship between input features and the target variable. The key difference is that KMeans is for unsupervised clustering, while KNN is for supervised classification or regression.

```python
# K-Means: unsupervised learning (no y)

number_clusters = 3

K_Means_obj = KMeans(n_clusters=number_clusters, random_state=0).fit(X)

labels= K_Means_obj.labels_
    
labels
```

Evaluating K-means clustering can be done using several metrics:
Inertia: Measures cluster compactness as the sum of squared distances within clusters. Lower values are better.
Silhouette Score: Assesses how well an object fits in its cluster compared to others, ranging from -1 to 1. Higher scores indicate better fit.
Davies-Bouldin Index: Indicates the average similarity between clusters, with lower values signifying better separation.
Calinski-Harabasz Index: Evaluates the ratio of between-cluster dispersion to within-cluster dispersion. Higher scores suggest denser, well-separated clusters.
Gap Statistic: Compares within-cluster variation to expected values under a null distribution, helping to identify the optimal number of clusters.
Cluster Stability: Measures consistency across multiple runs of the algorithm, with more stable results indicating better clustering.
These metrics help assess different aspects of clustering, like compactness, separation, and consistency, to determine the effectiveness of K-means on a dataset.
