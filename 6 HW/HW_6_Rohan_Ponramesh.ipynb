**Define the Mean Squared Error (MSE):**

The MSE for linear regression in 1D is given by:

$${\cal L}(m,b) = \frac{1}{2N} \sum_{d=1}^{N} (y_d - mx_d - b)^2$$


Where:
- $m$ is the slope of the linear regression line.
- $b$ is the intercept of the linear regression line.
- $N$ is the number of data points.
- $x_d$ is the independent variable (predictor) for the $d$-th data point.
- $y_d$ is the dependent variable (target) for the $d$-th data point.

**Minimization Condition:**

To find the optimal values of $m$ and $b$, we need to minimize the MSE with respect to $m$ and $b$. This means setting the partial derivatives with respect to $m$ and $b$ equal to zero.

**Partial Derivative with Respect to $m$:**

$$
\frac{\partial\mathcal{L}}{\partial m} = -\frac{1}{N} \sum_{d=1}^{N} x_d(y_d - mx_d - b)
$$

Setting this derivative to zero:

$$
0 = -\frac{1}{N} \sum_{d=1}^{N} x_d(y_d - mx_d - b)
$$

Simplifying:

$$
0 = \sum_{d=1}^{N} (x_dy_d - mx_d^2 - bx_d)
$$

**Partial Derivative with Respect to $b$:**

$$
\frac{\partial\mathcal{L}}{\partial b} = -\frac{1}{N} \sum_{d=1}^{N} (y_d - mx_d - b)
$$

Setting this derivative to zero:

$$
0 = -\frac{1}{N} \sum_{d=1}^{N} (y_d - mx_d - b)
$$

Simplifying:

$$
0 = \sum_{d=1}^{N} (y_d - mx_d - b)
$$

**Solve for** $m$ **and** $b$ **:**

We now have a system of two equations with two unknowns:

1. $\sum_{d=1}^{N} (x_dy_d - mx_d^2 - bx_d) = 0$
2. $\sum_{d=1}^{N} (y_d - mx_d - b) = 0$

Solve this system to find $m$ and $b$. To do this, we can rewrite these equations in terms of statistical quantities:

**Define Statistical Quantities:**
- $\mu_x$ is the mean of $x$.
- $\mu_y$ is the mean of $y$.
- $\mu_{xy}$ is the mean of $xy$.
- $\mu_{x^2}$ is the mean of $x^2$.

**Equations in Terms of Statistical Quantities:**

1. $\sum_{d=1}^{N} (x_dy_d - mx_d^2 - bx_d) = 0$ can be written as:
$$
N\mu_{xy} - mN\mu_{x^2} - bN\mu_x = 0
$$


2. $\sum_{d=1}^{N} (y_d - mx_d - b) = 0$ can be written as:
$$
N\mu_y - mN\mu_x - bN = 0
$$

**Solve for** $m$ **and** $b$ **:**

We have two equations and two unknowns:

1. $N\mu_{xy} - mN\mu_{x^2} - bN\mu_x = 0$
2. $N\mu_y - mN\mu_x - bN = 0$

Solving this system for $m$ and $b$, we get:

$$
m = \frac{\mu_{xy} - \mu_x\mu_y}{\mu_{x^2} - \mu_x^2}
$$

$$
b = \frac{\mu_{x^2}\mu_y - \mu_x\mu_{xy}}{\mu_{x^2} - \mu_x^2} 
$$

These equations give us the optimal values of the slope ($m$) and intercept ($b$) for the linear regression line in terms of statistical quantities (means, covariances, and variances) of the data.

These values of $m$ and $b$ minimize the mean squared error (MSE), providing the best-fit line to the data.

*In a markdown cell describe covariance and correlation, explain the difference and why we need two related quantities. That is, what does correlation do for us that covariance does not?*

ðŸ¤”

Covariance and correlation are two related statistical measures used to describe the relationship between two random variables in a dataset. While they are similar in some ways, they serve slightly different purposes and have different properties.

**Covariance:**

Covariance is a measure of how two variables change together. It indicates whether an increase in one variable corresponds to an increase or a decrease in another. The formula for calculating the covariance between two variables, X and Y, is as follows:

![Image](https://www.gstatic.com/education/formulas2/553212783/en/covariance_formula.svg)

Covariance can take on both positive and negative values. A positive covariance suggests that the two variables tend to increase together, while a negative covariance suggests that they tend to move in opposite directions. However, the magnitude of covariance is not standardized, making it difficult to interpret and compare across different datasets.

**Correlation:**

Correlation, on the other hand, is a standardized measure of the strength and direction of the linear relationship between two variables. It's a dimensionless value that falls within the range of -1 to 1, where:

- A correlation of 1 indicates a perfect positive linear relationship.
- A correlation of -1 indicates a perfect negative linear relationship.
- A correlation of 0 indicates no linear relationship.

The most commonly used method for calculating correlation is Pearson's correlation coefficient, which is given by the following formula:

![Image](https://www.gstatic.com/education/formulas2/553212783/en/correlation_coefficient_formula.svg)

**Differences and Why We Need Both:**

1. **Standardization:** The primary difference between covariance and correlation is that correlation is standardized, while covariance is not. Standardization means that correlation is independent of the units of measurement, making it easier to interpret and compare across different datasets.

2. **Interpretability:** Correlation provides a clear indication of the strength and direction of the linear relationship between two variables, ranging from -1 to 1. Covariance, in contrast, doesn't have a clear range and can be challenging to interpret without additional context.

3. **Comparability:** Correlation is more suitable for comparing relationships between different pairs of variables because it is standardized. Covariance's values depend on the variances of the variables, which makes it less suitable for comparisons.

While covariance measures the degree to which two variables change together, correlation takes this concept a step further by providing a standardized measure that is easier to interpret and compare across different datasets. Correlation is valuable because it quantifies the strength and direction of the linear relationship between variables and helps researchers and analysts make meaningful comparisons and draw more generalizable conclusions.

In a markdown cell, describe three of Tufte's rules of good visualization.

ðŸ¤”

>Edward Tufte is a renowned expert in the field of data visualization and has outlined several principles for creating effective and informative data visualizations. Tufte's rules emphasize the importance of clarity, simplicity, and truthfulness in data visualization. By adhering to these principles, data visualizations can effectively convey information to the audience and support better decision-making and understanding of complex data.

>Here are three of Tufte's rules of good visualization:
>1. **Optimize Data-Ink Utilization:**
   Tufte's principle of optimizing the utilization of data ink underscores the significance of removing non-essential ink or "chartjunk" from a visualization. It asserts that every pixel or ink mark in a graph or chart should serve the purpose of conveying data or information. Extraneous elements such as unnecessary gridlines, decorations, and embellishments should be eliminated to reduce clutter and amplify the clarity of the data being presented. This principle promotes simplicity and minimalism in data visualization, enabling information to stand out and be easily comprehensible.

>2. **Data Fidelity and Truthfulness:**
   Tufte places great emphasis on preserving the fidelity and truthfulness of data in visualizations. This entails accurately representing data without distorting or misrepresenting the information. It involves refraining from practices like altering axis scales, truncating data, or employing 3D effects to exaggerate differences between data points. Tufte advocates for a commitment to truth and honesty in data visualization to ensure that viewers can have confidence in the information being conveyed.

>3. **Sparklines:**
   Tufte introduced the concept of "sparklines" as a means to present concise, data-rich visualizations within the context of text or tables. Sparklines are miniature line charts or bar charts designed to provide a quick, at-a-glance representation of data trends. These compact visualizations are meant to be embedded in reports, tables, or documents, offering a visual summary of data without the need for a separate, larger chart. Sparklines are intended to be straightforward and information-dense, aiding in the comprehension of data trends and variations.

## <font color=#BB00FF>Problem #5 (5pts)</font>

Create a dataset that is a noisy line; that is, you have $y= mx+b$ and you want to add a controllable amount of noise to that. From the equations you derived in the first problem, use that data to compute the slope $m'$ and intercept $b'$, where the prime indicates a value inferred from the data.

Confirm with a plot that you are getting a good fit to your data. Vary the noise, make more plots and comment on how well you are able to predict the actual values of $m$ and $b$.

Finally, in a markdown cell, discuss how you could use the code you just wrote to handle missing values. Are there other steps you would want to add?


```python
import numpy as np
import matplotlib.pyplot as plt

m = 2.0
b = 3.0

x = np.linspace(0, 10, 100)

y_true = m * x + b

def add_noise(data, noise_level):
    return data + np.random.normal(0, noise_level, len(data))

noise_levels = [0.5, 1.0, 2.0]
datasets = [add_noise(y_true, noise) for noise in noise_levels]

fitted_lines = []
for dataset in datasets:
    m_prime = np.cov(x, dataset, bias=True)[0, 1] / np.var(x)
    b_prime = np.mean(dataset) - m_prime * np.mean(x)
    fitted_lines.append((m_prime, b_prime))

plt.figure(figsize=(12, 8))
for i, (dataset, (m_prime, b_prime)) in enumerate(zip(datasets, fitted_lines)):
    plt.subplot(3, 1, i + 1)
    plt.scatter(x, dataset, label="Noisy Data (Noise Level {:.1f})".format(noise_levels[i]))
    plt.plot(x, m * x + b, color='red', label="True Line (m={:.2f}, b={:.2f})".format(m, b))
    plt.plot(x, m_prime * x + b_prime, color='green', linestyle='--', label="Fitted Line (m'={:.2f}, b'={:.2f})".format(m_prime, b_prime))
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Noisy Line with Fitted Line (Noise Level {:.1f})'.format(noise_levels[i]))

plt.tight_layout()
plt.show()
```

**Handling Missing Value**

- While handling missing values, we can replace the noisy data with a dataset that includes NaN values where data is missing. 

- Then While calculating Mean and Covariance we can ignore NaN values.

- This will account for missing values while estimating the slope and intercept.

```python
missing_data = datasets[0].copy()
missing_data[20:40] = np.nan 

def nan_mean(data):
    return np.nanmean(data)

def nan_cov(x, y):
    mask = ~np.isnan(x) & ~np.isnan(y)
    return np.cov(x[mask], y[mask], bias=True)[0, 1]


fitted_lines = []
m_prime = nan_cov(x, missing_data) / nan_cov(x, x)
b_prime = nan_mean(missing_data) - m_prime * nan_mean(x)
fitted_lines.append((m_prime, b_prime))

plt.figure(figsize=(12, 8))
for i, (dataset, (m_prime, b_prime)) in enumerate(zip(datasets, fitted_lines)):
    plt.subplot(3, 1, i + 1)
    plt.scatter(x, dataset, label="Noisy Data (Noise Level {:.1f})".format(noise_levels[i]))
    plt.plot(x, m * x + b, color='red', label="True Line (m={:.2f}, b={:.2f})".format(m, b))
    plt.plot(x, m_prime * x + b_prime, color='green', linestyle='--', label="Fitted Line (m'={:.2f}, b'={:.2f})".format(m_prime, b_prime))
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Noisy Line with Fitted Line (Noise Level {:.1f})'.format(noise_levels[i]))

plt.tight_layout()
plt.show()
```
