## <font color=#11FF44> Fitting with the Moore-Penrose Pseudoinverse </font>

The second approach to computing the linear models for your data uses what is called the Moore-Penrose pseudoinverse (or, sometimes, just the pseudoinverse). We'll [cover it in two weeks](https://en.wikipedia.org/wiki/Moore%E2%80%93Penrose_inverse), but you can simply think of it as the inverse of a matrix that is not square. And, instead of the notation $A^{-1}$ we use $A^+$ to indicate pseudoinverse.

For today, I will just show you how to do this with code so that you can take the data from above and find the weights of the model.

The code below is pretty obvious, but what is does is make some noisy fake data and fit it to a line. From a math perspective, the idea is fairly simple. We take our dependent variables and put them into a vector ${\bf y}$; these are the Natural Growth Rate. You can take that column out of the dataframe and put it into an array. It will look like this:
$${\bf y} = \begin{bmatrix}y_1 \\ y_2 \\ \vdots  \\ y_n\end{bmatrix} .$$

Then, we form a matrix ${\bf A}$ with the _x_-values that go with those _y_ values, which looks like this:
$${\bf A} = \begin{bmatrix} 1 & x_1 \\ 1 & x_2 \\ \vdots & \vdots \\ 1 & x_n\end{bmatrix} .$$

We then have our regression equation
$$\begin{bmatrix}y_1 \\ y_2 \\ \vdots  \\ y_n\end{bmatrix} = \begin{bmatrix} 1 & x_1 \\ 1 & x_2 \\ \vdots & \vdots \\ 1 & x_n\end{bmatrix} \begin{bmatrix}w_1 \\ w_2 \end{bmatrix} .$$

Discuss with your group whether these vectors and matrix have consistent sizes/shapes. And, note that everything comes from the data columns in your dataframe, except the weights we want to our linear model. Open a markdown cell and explain all of this, including why we need a column with just ones in it. How is a "bias" being included?

Now, let's turn to the code that solves this equation and plot our solution. I'll make some fake data --  you need to understand the code and put the Growth Rate versus Age data in.

Note that the solution is **one line**: most of the effort is setting up the matrices and plotting the result! Note very carefully the use of the Python symbol "@". Research what this is and explain it in a markdown cell.



```python
import numpy as np
import matplotlib.pyplot as plt

num_points = 30
# fake data
x = np.linspace(0,60,num_points) + 2.0*np.random.randn(num_points)
y = x + np.random.randn(num_points)*2.0

# build matrices
ones = np.ones(num_points)
A = np.stack((ones,x)).transpose() # explain what this is doing

# solve equation - explain in detail what this code does
w_1, w_2 = np.linalg.pinv(A)@y

# plot results
x_for_model = np.linspace(0,60,100)
plt.plot(x_for_model, w_1 + w_2*x_for_model, label='model')
plt.plot(x, y, 'r+', label='data')
plt.legend()
plt.grid(alpha=0.16)
```

## <font color=#11FF44> Imputing Missing Values With Linear Model </font>

You are ready for the first step in imputing. Take the data with missing values and fit it to a line. (Make plots as you go so that you can confirm each step is working.) From the known _x_ values where the data is missing, create _y_ values using your model (that is, the line you found); save those points to plot the "non-stochastic" imputation - this is what you would use if you only used the linear model for imputation.

Next, subtract the _y_ values of the data from the model and find the standard deviation of the resulting values. Again using the known _x_ values where the data is missing, impute _y_ values using the line but now add random numbers using the standard deviation you just computed. (All of these operations are very simple using Numpy methods.)

Combine the imputed values you created with the dataset so that you can fit all of it together now.


```python
df_removed.head()
```

```python
df_test= df_removed.copy(deep=True)
df_test.head()
```

```python

df_test['Natural growth rate - Sex: all - Age: all - Variant: estimates'].loc[:]=np.NAN
df_test.head()
```

```python
def impute_func(row):
    
    x=row.loc['Median age - Sex: all - Age: all - Variant: estimates']
    #print(x)
    y=m_prime*float(x)+b_prime
    return y
```

```python
df_test['Natural growth rate - Sex: all - Age: all - Variant: estimates'] = df_test.apply(impute_func, axis=1)
df_test.head()
```

```python
g = sns.FacetGrid(df_removed, height = 6)
g = g.map(plt.scatter,  "Median age - Sex: all - Age: all - Variant: estimates","Natural growth rate - Sex: all - Age: all - Variant: estimates", edgecolor="w")
plt.plot(df_test["Median age - Sex: all - Age: all - Variant: estimates"],df_test["Natural growth rate - Sex: all - Age: all - Variant: estimates"], color='r')
plt.show()
```

```python
df_test["Natural growth rate - Sex: all - Age: all - Variant: estimates"]
```

```python
diff_imputed_original=df_test["Natural growth rate - Sex: all - Age: all - Variant: estimates"]-df_removed["Natural growth rate - Sex: all - Age: all - Variant: estimates"]
diff_imputed_original

std=np.std(diff_imputed_original)
std
```

```python
diff_imputed_original
```

```python
def impute_func_with_std(row):
    random_array = np.random.rand(-std, std)
    idx=np.random.randint(len(random_array)-1)
    
    x=row.loc['Median age - Sex: all - Age: all - Variant: estimates']
    #print(x)
    y=m_prime*float(x)+b_prime +random_array[idx]
    return y
```
